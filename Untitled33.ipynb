```r
install.packages(c("ggplot2", "tidyr", "dplyr"))

library(ggplot2)
library(tidyr)
library(dplyr)



df <- read.csv("Cardio.csv")
df <- data.frame(df)
colnames(df)  # Check column names
```

```r


df %>% glimpse()
summary(df)       #0 Basic stats
head(df)          # Top rows

data_clean <- df %>%
  clean_names() %>%                 # Column names ko snake_case mein convert karo
  drop_na() %>%                     # Remove rows with any NA
  filter(!is.na(income))            # Ensure 'income' column has no NA

# Replace with actual categorical column
data_clean %>%
  group_by(product) %>%
  summarise(
    avg_income = mean(income, na.rm = TRUE),
    total_miles = sum(miles, na.rm = TRUE),
    count = n()
  ) %>%
  arrange(desc(avg_income))

```

```r
options(scipen=999)
data_clean %>%
  ggplot(aes(x = product, y = income, fill = product)) +
  geom_boxplot() +
  theme_minimal() +
  labs(
    title = "Income Distribution by Product",
    x = "Product",
    y = "Income"
  )



data_clean %>%
  group_by(education) %>%
  summarise(avg_fitness = mean(fitness, na.rm = TRUE)) %>%
  ggplot(aes(x = education, y = avg_fitness, fill = education)) +
  geom_col() +
  labs(title = "Average Fitness by Education Level")
```
