```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Ankit_pal.xlsx")
df
```

# **Rename the column names**

```python
df = df.rename(columns={'How often do you like to visit this restaurant?': 'Visit_Frequency',
                        'How long did you wait for order?': 'Wait_Time',
                        'Who do you usually visit this restaurant with?':'Visit_With',
                        'Are you satisfied with the price rates of this restaurant?': 'Price_Satisfaction',
                        'What grade would you give this resturent?': 'Restaurant_Grade',
                        'How would you rate the behavior of staff/service?': 'Staff_Service_Rating',
                        'Would you recommended this restaurant to other?':'Recommendation'})

df["Wait_Time"] = df["Wait_Time"].replace({
    "20-30min": "20-30 min",
    "less than 10 min": "Less than 10 min"
})

df
```

# Remove whitespace from these columns

```python
df["Purpose"]=df["Purpose"].str.strip()
df["Gender"]=df["Gender"].str.strip()
df["Payment type"]=df["Payment type"].str.strip()
df["Visit_Frequency"]=df["Visit_Frequency"].str.strip()
df["Wait_Time"]=df["Wait_Time"].str.strip()
df["Age_group"]=df["Age_group"].str.strip()
df["Menu Card"]=df["Menu Card"].str.strip()
df["Visit_With"]=df["Visit_With"].str.strip()
df["Price_Satisfaction"]=df["Price_Satisfaction"].str.strip()
df["Restaurant_Grade"]=df["Restaurant_Grade"].str.strip()
df["Recommendation"]=df["Recommendation"].str.strip()
df
```

# **Restaurant Customer Feedback Dashboard**

```python
sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Restaurant Customer Feedback Dashboard", fontsize=18, fontweight="bold")
#plt.subplot(2,3,1)
sns.histplot(df["Rating"], bins=5, kde=True, color="blue", ax=axes[0,0])
axes[0,0].set_title("Customer Ratings Distribution")
axes[0,0].set_xlabel("Rating")
axes[0,0].set_ylabel("Count")

#plt.subplot(2,3,2)
sns.countplot(x="Payment type", data=df, palette="Set2", ax=axes[0,1],hue="Payment type")
axes[0,1].set_title("Preferred Payment Methods")
axes[0,1].set_xlabel("Payment Type")
axes[0,1].set_ylabel("Customers")

#plt.subplot(2,3,3)
sns.barplot(x="Age_group", y="Rating", data=df, palette="coolwarm", estimator="mean", ax=axes[0,2],hue="Age_group")
axes[0,2].set_title("Average Rating by Age Group")
axes[0,2].set_xlabel("Age Group")
axes[0,2].set_ylabel("Avg Rating")

#plt.subplot(2,3,4)
sns.countplot(x="Visit_Frequency", hue="Gender", data=df, palette="pastel", ax=axes[1,0])
axes[1,0].set_title("Visit Frequency by Gender")
axes[1,0].set_xlabel("Visit Frequency")
axes[1,0].set_ylabel("Customers")

#plt.subplot(2,3,5`)
sns.barplot(x="Wait_Time", y="Rating", data=df, hue="Gender" , ax=axes[1,1] ,palette=["Red","Green"])
axes[1,1].set_title("Wait Time vs Customer Rating")
axes[1,1].set_xlabel("Wait Time (minutes)")
axes[1,1].set_ylabel("Rating")

axes[1,2].axis("off")

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```
