# **Here's how you can fill missing values using the `.fillna()` method:**

```python
import pandas as pd
import numpy as np

# Recreate the DataFrame with missing values
data = {
    'Region': ['East', 'West', 'East', 'South', 'West', 'South'],
    'Product_Type': ['Electronics', None, 'Electronics', 'Electronics', 'Clothing', 'Clothing'],
    'Customer_Segment': ['Regular', 'Premium', 'Regular', 'Premium', 'Regular', 'Premium'],
    'Sales': [1000, 1500, 1200, 2000, 1800, 1300],
    'Quantity': [5, np.nan, 7, 12, 8, np.nan]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Fill numerical missing values in 'Quantity' with the mean of the column
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())

# Fill numerical missing values in 'Sales' with the mean of the column
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())


# Fill categorical missing values (in 'Product_Type') with a placeholder string or mode
# Let's fill with a placeholder 'Unknown'
df['Product_Type'] = df['Product_Type'].fillna('Unknown')

print("\nDataFrame after filling missing values:")
print(df)

# You can also fill all missing values in the entire DataFrame with a single value
# Be careful with this if your DataFrame has mixed data types
# df_filled_with_zero = df.fillna(0)
# print("\nDataFrame after filling all missing values with 0:")
# print(df_filled_with_zero)
```

# **joining and merging of 2  data frames**

```python
import pandas as pd

# Sample DataFrame 1
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['Alice', 'Bob', 'Charlie', 'David']}
df1 = pd.DataFrame(data1)

# Sample DataFrame 2
data2 = {'ID': [1, 2, 3, 5],
         'Salary': [60000, 75000, 50000, 90000]}
df2 = pd.DataFrame(data2)

print("DataFrame 1:")
print(df1)


print("\nDataFrame 2:")
print(df2)

merged_df_inner = pd.merge(df1, df2, on='ID', how='inner')
print("\ninner (keep all common from df1,df2):")
print(merged_df_inner)

#left merge (keep all rows from df1, add matching rows from df2)
merged_df_left = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Merge (keep all from df1):")
print(merged_df_left)

'''
# Right merge (keep all rows from df2, add matching rows from df1)
merged_df_right = pd.merge(df1, df2, on='ID', how='right')
print("\nRight Merge (keep all from df2):")
print(merged_df_right)
'''
'''
# Outer merge (keep all rows from both dataframes)
merged_df_outer = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Merge (keep all rows):")
print(merged_df_outer)
'''

# --- Joining DataFrames ---
# Join is similar to merge, often used when joining on index.
# Let's set 'ID' as the index for joining example
df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')

print("\nDataFrame 1 with ID as index:")
print(df1_indexed)

print("\nDataFrame 2 with ID as index:")
print(df2_indexed)

# Inner join on index
joined_df_inner = df1_indexed.join(df2_indexed, how='inner')
print("\nInner Join (matching index only):")
print(joined_df_inner)

# Left join on index
joined_df_left = df1_indexed.join(df2_indexed, how='left')
print("\nLeft Join (keep all from df1_indexed):")
print(joined_df_left)

# Right join on index
joined_df_right = df1_indexed.join(df2_indexed, how='right')
print("\nRight Join (keep all from df2_indexed):")
print(joined_df_right)

# Outer join on index
joined_df_outer = df1_indexed.join(df2_indexed, how='outer')
print("\nOuter Join (keep all index values):")
print(joined_df_outer)
joined_df_cross = df1_indexed.join(df2_indexed, how='cross')
joined_df_cross
```

**Reasoning**:
Calculate the total quantity sold for each product and the average price of products sold in each city, which addresses the first two intermediate level questions.



```python
# What is the total quantity sold for each product?
total_quantity_per_product = df.groupby('Product')['Quantity'].sum()
print("Total quantity sold for each product:")
display(total_quantity_per_product)

# What is the average price of products sold in each city?
average_price_per_city = df.groupby('City')['Price'].mean()
print("\nAverage price of products sold in each city:")
display(average_price_per_city)
```

**Reasoning**:
Calculate the total sales per Manager, extract the month from the Date column, find the most common payment method, count how many times each product was sold, and find orders with Quantity greater than 5 and Purchase Type "Online", which addresses the remaining intermediate level questions.



```python
# Find the total sales (Price * Quantity) per Manager.
df['Total Sale'] = df['Price'] * df['Quantity']
total_sales_per_manager = df.groupby('Manager')['Total Sale'].sum()
print("\nTotal sales per Manager:")
display(total_sales_per_manager)

# Extract the month from the Date column and create a new column called Month.
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Month'] = df['Date'].dt.month
print("\nDataFrame with 'Month' column:")
display(df.head())

# Find the most common payment method.
most_common_payment_method = df['Payment Method'].value_counts().idxmax()
print("\nMost common payment method:", most_common_payment_method)

# How many times was each product sold?
product_sales_count = df['Product'].value_counts()
print("\nNumber of times each product was sold:")
display(product_sales_count)

# Find all orders made with a Quantity greater than 5 and a Purchase Type of "Online".
online_orders_high_quantity = df[(df['Quantity'] > 5) & (df['Purchase Type'] == 'Online')]
print("\nOnline orders with Quantity greater than 5:")
display(online_orders_high_quantity)
```
