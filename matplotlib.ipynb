```python
import matplotlib.pyplot as plt
import pandas as pd

# Identify numerical columns
numerical_cols = df_cardio.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, 10))
plt.boxplot(df_cardio[numerical_cols].values, tick_labels=numerical_cols)
plt.title('Boxplots of Numerical Columns')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df_cardio['Age'],  bins=10,edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y',alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Age vs Income
plt.figure(figsize=(10, 6))
plt.scatter(df_cardio['Age'], df_cardio['Income'])
plt.title('Scatter Plot of Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart of Gender distribution
gender_counts = df_cardio['Gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Gender')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(gender_counts.index)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df_cardio['Education'])
plt.title('Boxplot of Education')
plt.ylabel('Education Level')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Bar chart of average Income by Product
income_by_product = df_cardio.groupby('Product')['Income'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(income_by_product['Product'], income_by_product['Income'])
plt.title('Average Income by Product')
plt.xlabel('Product')
plt.ylabel('Average Income')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate the average income per education level
income_by_education = df_cardio.groupby('Education')['Income'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(income_by_education['Education'], income_by_education['Income'], marker='o', linestyle='-')
plt.title('Average Income by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.grid(True)
plt.show()
```

Understanding which columns are suitable for different types of graphs is key to effective data visualization. Here's a breakdown:

**Numerical Columns (e.g., Age, Income, Miles, Education, Usage, Fitness):** These columns contain numerical values that can be measured.

*   **Histograms:** Suitable for showing the distribution of a single numerical variable. For example, a histogram of the **'Age'** column shows the distribution of ages in the dataset.
*   **Boxplots:** Useful for visualizing the distribution, central tendency, and variability of a numerical variable. For example, a boxplot of the **'Income'** column shows the distribution of income and potential outliers. Boxplots are also great for comparing the distribution of a numerical variable across different categories, like plotting **'Income'** for each **'Product'**.
*   **Scatter Plots:** Used to show the relationship between two numerical variables. For example, a scatter plot of **'Age'** vs **'Income'** can show if there is a correlation between age and income.
*   **Line Charts:** Can be used to show trends in a numerical variable, especially when there's an ordered category or time series data. In this dataset, we used it to show the trend of average **'Income'** across **'Education'** levels.

**Categorical Columns (e.g., Product, Gender, MaritalStatus):** These columns contain values that represent categories or groups.

*   **Pie Charts:** Suitable for showing the proportion or percentage of each category within a single categorical variable. For example, a pie chart of the **'Gender'** column shows the proportion of male and female customers. They are best used when you have a small number of categories.
*   **Bar Charts:** Used to compare the values of a numerical variable across different categories or to show the frequency of each category in a categorical variable. We used it to compare the average **'Income'** across different **'Product'** types. You could also use a bar chart to show the count of each **'MaritalStatus'**.

**In summary:**

*   Use **histograms and boxplots** for exploring the distribution of single numerical variables (**'Age'**, **'Income'**, etc.).
*   Use **scatter plots** to examine the relationship between two numerical variables (**'Age'** vs **'Income'**, **'Usage'** vs **'Miles'**, etc.).
*   Use **line charts** for trends, especially with ordered categories (**'Education'** vs **'Income'**).
*   Use **pie charts and bar charts** for visualizing categorical data (**'Gender'**, **'MaritalStatus'**) and comparing numerical data across categories (**'Income'** by **'Product'**).
