```python
import pandas as pd
import requests

# Step 1: Choose a reliable URL with a population table
url = "https://www.worldometers.info/population/most-populous-countries/"

# Step 2: Use requests to fetch the page with headers
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Step 3: Parse tables from the HTML content
tables = pd.read_html(response.text)

# Step 4: Display the first table (usually the main one)
df = tables[2]
print(df)
# Remove columns with any NaN values
df_cleaned = df.dropna(axis=1)

# Display the cleaned DataFrame
display(df_cleaned)
```
