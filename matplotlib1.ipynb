```python
#line chart
import matplotlib.pyplot as plt
import numpy as np

name=['a','b','c','d']
value=[2,3,6,5]
plt.plot(name,value,'*-.r')
plt.xlabel('name',size=20,c='cyan')
plt.ylabel('value',size=20,c='cyan')
plt.title('line chart',size=25,c='m')
plt.grid(axis='y',lw=2.5,c='g')
plt.show()

#bar chart
import matplotlib.pyplot as plt
import numpy as np

name=['a','b','c','d']
value=[2,3,6,5]
plt.bar(name,value,color='r',labels="bar")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'sales': [100, 150, 120, 180, 200, 170],
        'year':[1,2,3,4,5,6]}
df = pd.DataFrame(data)
display(df)
colors=np.random.randint(0,100,6)
plt.scatter(df['year'],df['sales'],alpha=0.5,s=20,cmap='viridis',c=colors)
plt.xlabel('year')
plt.ylabel('sales')
plt.colorbar()
plt.title('scatter plot')
plt.show()


plt.pie(df['sales'],labels=df['month'],autopct='%1.2f%%',explode=[0.2,0,0,0,0,0.5],shadow=True,startangle=90)
plt.legend(title='month')

plt.legend(loc='lower right', ncol=6, bbox_to_anchor=(.25, .15))


plt.figsize=(640,480)
plt.xlabel('month')
plt.ylabel('sales')
plt.title('pie chart')
plt.show()
plt.savefig("salarypie.jpeg")
```
